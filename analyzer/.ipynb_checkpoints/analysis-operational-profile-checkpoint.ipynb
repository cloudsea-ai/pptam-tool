```r
# Analysis of the Operational Profile

## Libraries
```

```r
install.packages("RCurl", dependencies = TRUE)
library(RCurl)
library(repr)
```

## 1. Loading the Operational Profile from Wikipedia

```r
### Options

profile.year <- 2016
profile.month <- 7
profile.min.day <- 1
profile.max.day <- 31

profile.project <- "de"
profile.filename <- paste("wikipedia-profile-", profile.project, ".csv", sep = "")

### END Options

read.project.counts <- function(year, month, day, hour) {
    URL <- paste("https://dumps.wikimedia.org/other/pagecounts-raw/",
                 sprintf("%04d/", year), sprintf("%04d-%02d", year, month),
                 sprintf("/projectcounts-%04d%02d%02d-%02d0000", year, month, day, hour),
                 sep = "")
    
    # print(paste("Fetching", URL, "..."))
    
    read.csv2(textConnection(getURL(URL)),
              header = FALSE, sep = " ",
              col.names = c("project", "page", "requests", "size"))
}

project.count.data <- data.frame(date = c(), requests = c())
i <- 1

for (day in profile.min.day:profile.max.day) {
    for (hour in 0:23) {
        project.counts <- read.project.counts(profile.year, profile.month, day, hour)
        num.requests <- project.counts[project.counts$project == profile.project, "requests"]
        
        project.count.data <- rbind(project.count.data,
                                    data.frame(date = ISOdate(profile.year, profile.month, day, hour),
                                               requests = num.requests))
        
        i <- i + 1
    }
}

write.csv2(project.count.data, file = profile.filename, row.names = FALSE)
```

## 2. Visualizing the Operational Profile

```r
## Uncomment this if you want to skip the first step
# project.count.data <- read.csv2(profile.filename)
# profile.filename <- paste("wikipedia-profile-", profile.project, ".csv", sep = "")

options(repr.plot.width=10, repr.plot.height=6)
plot(project.count.data, type = "l", col = "blue", lwd = 1.5)
```

## 3. Determining the Load Levels

```r
## Options

num.load.levels <- 25

## END Options

options(repr.plot.width=10, repr.plot.height=6)
request.hist <- hist(project.count.data$requests,
     breaks = num.load.levels,
     col = "darkgray", border = "white",
     xlab = "requests", main = "Histogram of requests per hour")
```

## 4. Scale to the Maximum Number of Users

```r
## Options

max.num.users <- 300

## END Options


max.requests <- max(project.count.data$requests)
scale.factor <- max.num.users / max.requests
print(paste("Maximum is", max.requests, "requests, scale factor is", scale.factor))

scaled.number.of.users <- scale.factor * project.count.data$requests

options(repr.plot.width=10, repr.plot.height=6)
num.users.hist <- hist(scaled.number.of.users,
     breaks = num.load.levels,
     col = "darkgray", border = "white",
     xlab = "scaled number of users", main = "Histogram of scaled number of users")
```

## 5. Store the Frequency of Occurrence of each Load Level

```r
## Options

frequencies.of.occurrence.filename <- paste("wikipedia-load-frequencies-", profile.project, ".csv", sep = "")

## END Options

num.load.levels <- sum(num.users.hist$counts)
frequencies.of.occurrence <- data.frame(
    requests.per.hour = num.users.hist$mids / scale.factor,
    scaled.number.of.users = num.users.hist$mids,
    frequency.of.occurrence = num.users.hist$counts / num.load.levels
)

write.csv2(frequencies.of.occurrence, file = frequencies.of.occurrence.filename, row.names = FALSE)
print(paste("Wrote frequencies of occurrence to", frequencies.of.occurrence.filename))

frequencies.of.occurrence
```
