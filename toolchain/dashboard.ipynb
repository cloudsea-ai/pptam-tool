```python

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_json("./exported/polygons.json", orient='records')

fig = plt.figure(figsize=(10,5))
p = fig.add_subplot()
p.plot(pd.unique(data.x), data[data.group=="operational_profile"].relative_frequency, color='black', marker='o', label = "Operational Profile")
plt.title('Absolute mass', fontsize=14)

cmap = plt.get_cmap("tab10")
color = 0
for id, group in data[data.group!="operational_profile"].groupby("group"):
    p.plot(group.x, group.absolute_mass, color=cmap(color), marker='x', label = id)
    p.fill_between(group.x, group.absolute_mass, step=None, color=cmap(color), alpha=0.4)
    color += 1

p.set_xlabel('workload')
p.set_ylabel('frequency')
p.grid(True)
p.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
p = fig.add_subplot()
plt.title('Relative mass', fontsize=14)

cmap = plt.get_cmap("tab10")
color = 0
for id, group in data[data.group!="operational_profile"].groupby("group"):
    p.plot(group.x, group.relative_mass, color=cmap(color), marker='x', label = id)
    p.fill_between(group.x, group.relative_mass, step=None, color=cmap(color), alpha=0.4)
    color += 1

p.set_xlabel('workload')
p.set_ylabel('frequency')
p.grid(True)
p.legend()
plt.show()
```
